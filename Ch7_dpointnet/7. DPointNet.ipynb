# 7. Optimizing Networks with DPointNet

DPointNet is a built-in BMTK simulation engine that uses deep-learning techniques to optimize brain circuit models. It can be used both to run simulations of a network (or in machine-learning parlance running an **inference** on a model) as well as to **train** the weighted synaptic connections of the model. Like with the rest of BMTK, it can directly use SONATA files containing information about networks, configurations, and spikes/voltage recordings, thus allowing easy integration with other parts of the API like the **Network Builder**, **PointNet**, and **FIlterNet**. (It also has support for other formats: pickle files, npy files, NWB files, etc.).

DPointNet resulted from collaboration between the Allen Institute and the Wolfgang Maass lab at Graz University of Technology and Claudio Mirasso's lab at the University of Balearic Islands. For more information about the underlying science and motivation please refer to the following:

```
[1] 2026 V1 model: Ito, S., Haufler, D., Fraile, J. G., Dai, K., Aman, J., Chen, G., Mirasso, C., Maass, W., & Arkhipov, A. (2026). Deep-learning-assisted simulation of a cortical circuit: integrating anatomy, physiology and function. bioRxiv, 2026.03.13.711751. https://doi.org/10.64898/2026.03.13.711751
```
